# Desarrollo de curvas de calidad de estación por el método de la curva guía: modelo de Hossfeld I:

El objetivo de este notebook es obtener unas curvas de calidad de estación mediante el empleo del método de la curva guía: modelo de Hossfeld I $[H_0=\frac{t^2}{(a+bt)^2}]$. Para ello se van a emplear datos reales procedentes del análisis de tronco de 161 árboles de la especie Pinus radiata D. Don en Galicia. 
Para la realización de este ejercicio se van a usar las siguientes librerias: numpy, pandas, matplotlib y statsmodels.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from nls import NLS #nonlinear regression
```

El segundo paso será importar los datos del fichero datos1.txt a un dataframe.

```python
f00 =pd.read_csv('datos1.txt', sep='\t', thousands=',')
f00.head()
```

Una vez importados los datos y antes de proceder al ajuste por mínimos cuadrados es conveniente visualizar los datos en un gráfico de
dispersión en el que se represente en el eje de ordenadas la altura dominante y en el
eje de abscisas la edad.


```python
plt.figure(figsize=(7,5))
plt.plot(f00.edad,f00.altura,'o',color='grey')
plt.title("Gr"+u'á'+"fico de altura frente a edad")
plt.xlabel("Edad t (a"+u'ñ'+"os)")
plt.ylabel("Altura dominante $H_0$ (m)")
plt.show()
```

Las curvas de calidad de estación deben cumplir una serie de propiedades: tener una asíntota horizontal y un punto de inflexión, tener un comportamiento lógico (por ejemplo, pasar por el punto edad = 0, altura dominante = 0) y, además, la familia de curvas debe ser, salvo contadas ocasiones, polimórfica, es decir, las curvas no deben ser proporcionales. En esta notebook se va a emplear una simplificación del modelo matemático de Hossfeld I, que cumple las tres primeras propiedades descritas. Su expresión matemática es: $H_0=\frac{t^2}{(a+bt)^2}$

#Ajuste por regresión lineal
Para poder realizar el ajuste de regresión lineal se debe crear
previamente la variable dependiente, cuyos valores se obtienen como el cociente entre la edad y
la raíz cuadrada de la altura dominante.

```python
f00["vardep"]=f00.edad/f00.altura**0.5
```

A continuación procedemos a realizar el ajuste de regresión lineal:

```python
X = sm.add_constant(f00["edad"],prepend=True)
mod = sm.OLS(f00.vardep,X)
res = mod.fit()
res.summary()
```

Al realizar el ajuste de regresión lineal se obtiene un valor del coeficiente de
determinación R-squared = 0,864. También se comprueba que el modelo tiene sentido, puesto
que la probabilidad de equivocarse si se rechaza la hipótesis nula de que todos los
parámetros que acompañan a variables independientes (en este caso sólo el
parámetro b) son iguales a cero es muy baja (p-value asociado al F-statistic que aparece
al final del resumen del análisis de regresión lineal). El límite que se utiliza para
rechazar la hipótesis nula indicada depende del objetivo de la modelización, pero
generalmente se admite que se rechaza la hipótesis nula si la probabilidad es inferior a
0.05 (un 5%). Por último, se obtienen los valores de las estimaciones de los parámetros
a = 1,8594 (en la salida del programa se denomina Intercept por no acompañar a ninguna
variable independiente) y b = 0,1408 (en la salida del programa se denomina edad
puesto que es ésta la variable independiente a la que acompaña) y se comprueba que,
en la población, las estimaciones de esos parámetros no deben ser iguales a cero
puesto que las dos probabilidades de equivocarse si se rechazan las hipótesis nulas de
que cada uno de ellos es igual a cero son muy bajas (valores de la
columna Pr(>|t|)). Asumiendo nuevamente que el límite fijado para rechazar la
hipótesis nula anterior es 0.05, se dice que las estimaciones de los parámetros son
significativamente distintas de cero a un nivel del 5%.


#Ajuste por regresión no lineal
El siguiente paso es realizar el ajuste del modelo original por regresión no lineal
empleando como valores de partida del proceso iterativo para a y b los obtenidos en el
ajuste lineal.


```python
def hossfeld(params, t, yObs):
    a,b = params[0],params[1]            
    h = t**2/(a+b*t)**2 
    return(yObs - h)

nlmod = NLS(hossfeld,{'a':res.params["const"], 'b':res.params["edad"]}, f00['edad'], f00['altura'] )
nlmod.summary()
nlmod.AIC()

```

Los valores estimados son a = 1,779
y b = 0,1412. Se comprueba que las estimaciones de los parámetros son
significativamente distintas de cero a un nivel del 5% (tabla Parameters, valores de la
columna Pr(>|t|)). Por tanto, la curva guía tiene la siguiente expresión:
    
$H_0=\frac{t^2}{(1.779+0,1412·t)^2}$

Se puede comprobar gráficamente que la diferencia entre la curva media obtenida con
las estimaciones del modelo ajustado por regresión lineal y la obtenida con las del
ajustado por regresión no lineal es pequeña.


```python
df = pd.DataFrame({'edad':np.arange(1,42)})
df["H0pred_0"]= df["edad"]**2/(res.params["const"]+res.params["edad"]*df["edad"])**2

res.params["const"] = nlmod.parmEsts[0]
res.params["edad"] = nlmod.parmEsts[1]


df["H0pred"]= df["edad"]**2/(res.params["const"]+res.params["edad"]*df["edad"])**2

plt.figure(figsize=(10,8))
for name, group in f00.groupby('arbol'):
    plt.plot(group.edad,group.altura,color='grey', alpha = 0.4)
plt.plot(df.edad,df.H0pred,linewidth=1,color='red')
plt.plot(df.edad,df.H0pred_0,'--',linewidth=1,color='blue')
plt.text(3, 32, r"$H_0=\frac{t^2}{(%.3f+%.3ft)^2}$" % ( res.params["const"],res.params["edad"],), fontsize=20)
plt.title("Gr"+u'á'+"fico de la curva gu"+u'í'+"a por regresion lineal(rojo) y no lineal (azul)")
plt.xlabel("Edad t (a"+u'ñ'+"os)")
plt.ylabel("Altura dominante $H_0$ (m)")
plt.axis([-1,41,-1,41])
plt.show()
```

#Obtener la serie de curvas de calidad de estación 
Por último, una vez obtenida la curva media o curva guía, se deben representar la serie de curvas de calidad de estación de cada una de las familias que se pueden
obtener a partir de dicha curva guía. Para ello, se fija la edad de referencia en años (aproximadamente la mitad del turno según los datos de la gráfica altura
dominante-edad) y se decide que la calidad superior a representar tenga un índice sitio IS de 26 m, es decir, un valor de 26 m de altura dominante a esa edad de referencia, y que la calidad inferior tenga un índice de sitio de 11 m. Entre estas calidades extremas se decide representar otras dos calidades con intervalos de 5 entre ellas, es decir, con índices de sitio de 16 y 21 m. Teniendo en cuenta estas condiciones, la ecuación de la mejor calidad deberá cumplir la siguiente relación:
    
${IS}=\frac{t_{ref}^2}{(a+b·t_{ref}})^2 =26=\frac{20^2}{(a+b·20})^2$

Esta ecuación con dos incógnitas sólo se puede resolver si se le da un valor a unolos parámetros (a o b). El valor de dicho parámetro se obtiene de los valores estimados
para la curva guía y, por tanto, hay dos alternativas posibles:

1. Fijar b para todas las curvas de la familia y despejar a.

2. Fijar a para todas las curvas de la familia y despejar b.

En el primer caso se asume que el parámetro a es específico de cada estación y el valor
${a}_{is}$ de la curva correspondiente se obtiene como:

${IS}=\frac{t_{ref}^2}{({a}_{is}+ 0,141·t_{ref}})^2\Rightarrow{a}_{is}=\frac{t_{ref}}{ \sqrt{IS}} - 0,141·t_{ref}$
   
En el segundo caso se asume que el parámetro b es específico de cada estación yvalor ${b}_{is}$ de la curva correspondiente se obtiene como:

${IS}=\frac{t_{ref}^2}{(1,859+{b}_{is}·t_{ref}})^2\Rightarrow{b}_{is}=(\frac{t_{ref}}{ \sqrt{IS}} -  1,859)/t_{ref}$




El código de Python para generar las diferentes curvas puede ser el siguiente:

```python
edad_ref = 20
S = np.array([11, 16, 21, 26])
ds = pd.DataFrame({'s':S})
ds['a'] = ((edad_ref / ds['s'] ** 0.5)- res.params["edad"] * edad_ref)
ds['b'] = ((edad_ref / ds['s'] ** 0.5)- res.params["const"]) / edad_ref
for i in ds.index:  
    df['a_is' + str(ds.s[i])] = df.edad ** 2/(ds.a[i] + res.params["edad"] * df.edad) ** 2
    df['b_is' + str(ds.s[i])] = df.edad ** 2/(res.params["const"] + ds.b[i] * df.edad) ** 2
```

# Hossfeld I a

```python
from IPython.display import Latex
st = ""
for i in ds.index:
    st += r"""${IS}_{%s}=\frac{t^2}{(%.3f+%.3ft)^2} || $"""% (str(ds['s'][i]),ds['a'][i],res.params["edad"])
Latex(st)
```

# Hossfeld I b

```python
st = ""
for i in ds.index:
    st += r"""${IS}_{%s}=\frac{t^2}{(%.3f+%.3ft)^2} || $"""% (str(ds['s'][i]),res.params["const"],ds['b'][i])
Latex(st)
```

Las dos familias de curvas obtenidas son polimórficas. La diferencia estriba en que puesto que la asíntota horizontal del modelo es
$ \displaystyle\lim_{t \to\infty}\frac{t^2}{(a+bt)^2}=\frac{1}{b^2}$
 en la primera familia la asíntota es común y en la segunda familia cada curva tiene una asíntota
diferente. Analizando las dos familias de curvas sobre los datos originales se puede
apreciar que la familia con una única asíntota (izquierda, valor de b común) es la más
adecuada.


```python
plt.figure(figsize=(15,7),dpi=300)

plt.subplot(122)
for name, group in f00.groupby('arbol'):
    plt.plot(group.edad,group.altura,color='grey', alpha=0.3)
for i in ds.index:
    plt.plot(df.edad,df['b_is' + str(ds.s[i])],linewidth=2,color='black')
    plt.text(df.edad[36],df['b_is' + str(ds.s[i])][36],'IS-'+ str(ds.s[i]),rotation=0, color="black", fontsize=10,backgroundcolor = "white")
plt.title("Curva gu"+u'í'+"a Hossfeld I b")
plt.xlabel("Edad t (a"+u'ñ'+"os)")
plt.axis([-1,41,-1,45])

plt.subplot(121)
for name, group in f00.groupby('arbol'):
    plt.plot(group.edad,group.altura,color='grey', alpha=0.3)
for i in ds.index:
    plt.plot(df.edad,df['a_is' + str(ds.s[i])],linewidth=2,color='black')
    plt.text(df.edad[36],df['a_is' + str(ds.s[i])][36],'IS-'+ str(ds.s[i]),rotation=0, color="black", fontsize=10,backgroundcolor = "white")
plt.title("Curva gu"+u'í'+"a Hossfeld I a")
plt.xlabel("Edad t (a"+u'ñ'+"os)")
plt.ylabel("Altura dominante $H_0$ (m)")
plt.axis([-1,41,-1,45])
#savefig('grafica.png', dpi=300)
plt.show()
```
